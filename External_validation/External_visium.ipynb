```python
import scanpy as sc
Path = "/scratch/project/stseq/Onkar/HnC_ExternalVal/"
```

```python
Visium_S01_MAR21 = sc.read_visium(Path+"Visium_S01_MAR21/Visium_S01_MAR21/")
Visium_S01_MAR21.var_names_make_unique()
sc.pp.normalize_total(Visium_S01_MAR21, inplace=True)
sc.pp.log1p(Visium_S01_MAR21)
```

```python
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap
import scanpy as sc

# Define a custom grey to red colormap
grey_red_cmap = LinearSegmentedColormap.from_list(
    "grey_red",
    ["#D3D3D3", "#FFAA33","#FF0000"]  # light grey to red
)

# Example usage in Scanpy (e.g., with sc.pl.umap)
# adata: your AnnData object
# 'some_gene' is a gene name or a column in adata.obs
# sc.pl.umap(adata, color='some_gene', cmap=grey_red_cmap)
```

```python
sc.pl.spatial(Visium_S01_MAR21,color=["FMO2", "PAEP", "MARCO","PNPLA3"],spot_size=100,cmap=grey_red_cmap)
```

```python
sc.tl.score_genes(Visium_S01_MAR21,["FMO2", "PAEP", "MARCO","PNPLA3"])
Visium_S01_MAR21.obs["four_genes_add_score"] = Visium_S01_MAR21.to_df()[["FMO2", "PAEP", "MARCO","PNPLA3"]].sum(axis=1)
sc.pl.spatial(Visium_S01_MAR21,color=["VEGFA", "four_genes_add_score", "score"],spot_size=100,cmap=grey_red_cmap,vmin=0)
```
